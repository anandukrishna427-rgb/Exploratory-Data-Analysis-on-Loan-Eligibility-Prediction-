This dataset, sourced from Kaggle, is tailored for predicting loan approval eligibility based on applicants' demographic, financial, and credit profiles. It features 614 entries (train split) with 13 columns for binary classification tasks (Loan_Status: 'Y' for approved, 'N' for rejected).
Key columns include:

Customer_ID: Unique applicant identifier.
Gender, Married, Dependents, Education, Self_Employed: Demographic and employment details (categorical).
Applicant_Income, Coapplicant_Income: Income metrics (numerical).
Loan_Amount, Loan_Amount_Term: Loan specifics (numerical).
Credit_History: Binary credit score (1 for good, 0 for poor).
Property_Area: Location category (Urban, Semiurban, Rural).

Perfect for exploratory data analysis (EDA), handling imbalances/missing values, feature engineering (e.g., total income), and training ML models like logistic regression, decision trees, or random forests to evaluate lending risks and biases in financial decision-making.

# Feature Enginerring 

```python
df['emi amount'] = df['loan amount'] / df['loan amount term']
df['income loan ratio'] = df['applicant income'] / df['loan amount']
df['debt income ratio'] = df['emi amount'] / (df['applicant income'] / 12)
```

# Outlier Analysis

```python
import pandas as pd

numeric_cols = df.select_dtypes(include=['int64','float64']).columns

outliers = pd.DataFrame()

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    outliers[col] = df[(df[col] < lower) | (df[col] > upper)][col]

n = len(numeric_cols)
rows = (n // 3) + 1
cols = 3
fig, axes = plt.subplots(rows, cols, figsize=(15, 5 * rows))
axes = axes.flatten()
for i, col in enumerate(numeric_cols):
    sns.boxplot(x=df[col], ax=axes[i], color="skyblue")
    axes[i].set_title(f"Boxplot of {col}")
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()

```

# Conclusion

#### This exploratory data analysis of the loan approval dataset examined applicant demographics, financial attributes, and engineered affordability metrics to understand patterns influencing loan approval decisions. The analysis highlighted income stability, credit history, and repayment burden as key drivers of loan outcomes. Approved loans were commonly associated with lower EMI values, stronger income-to-loan ratios, and reduced debt-to-income stress, while rejected loans showed higher repayment pressure and weaker affordability indicators. Credit history emerged as a strong discriminator, with applicants having a positive credit record experiencing higher approval rates. Property location and education also showed moderate influence, while gender and marital status displayed minimal impact. Overall, the findings emphasize that repayment capacity and credit behavior play a central role in loan approval decisions, with the analysis conducted on the given loan dataset.

1. The dataset comprised applicant demographic, income, loan, and credit-related variables, along with engineered features such as EMI, income-loan ratio, and debt-income ratio to capture repayment behavior.
2. Approved loans generally exhibited lower EMI values, indicating manageable monthly repayment obligations.
3. Rejected loans showed higher and more variable debt-to-income ratios, reflecting increased repayment stress.
4. Applicants with a positive credit history had significantly higher loan approval rates compared to those without credit history.
5. Higher applicant income relative to loan amount was associated with improved approval likelihood.
6. Property area and education level showed moderate variation across loan status but were not primary approval drivers.
7. Demographic attributes such as gender and marital status had minimal influence on loan approval outcomes compared to financial and credit factors.
